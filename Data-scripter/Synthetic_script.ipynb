```python
import openai
import random

openai.api_key = "your-api-key"

def auto_generate_bfsi_service():
    # Predefined categories
    services = ["Policybazaar", "HDFC ERGO", "Aditya Birla Sun Life", "Kotak Mahindra", "ICICI Prudential"]
    domains = ["BFSI > Insurance", "BFSI > Banking", "BFSI > Investments"]
    platforms = ["Web", "Mobile", "Desktop"]
    focus_areas = ["Frontend Development", "Backend Development", "Mobile Application Development", "Full Stack Development", "DevOPS", ]
    
    # Functionality templates
    functionality_templates = {
        "Insurance": [
            "Policy Search & Comparison",
            "Claim Assistance",
            "Instant Renewal",
            "Premium Calculator",
            "Hospital & Garage Locator"
        ],
        "Banking": [
            "Fund Transfers",
            "Account Management",
            "Loan Applications",
            "Credit Card Services",
            "Bill Payments"
        ],
        "Investments": [
            "Portfolio Tracking",
            "Investment Recommendations",
            "Market Analysis",
            "Mutual Fund Purchase",
            "Stock Trading"
        ]
    }

    # Non-functionality templates
    non_functionality_templates = [
        "Performance",
        "Security",
        "Scalability",
        "Availability",
        "Accessibility"
    ]
    
    # Randomly select service and domain
    service_name = random.choice(services)
    domain = random.choice(domains)
    
    # Determine functionalities based on domain
    domain_key = domain.split(">")[-1].strip()
    functionalities = random.sample(functionality_templates.get(domain_key, []), k=3)
    
    # Generate other fields
    platform = random.choice(platforms)
    focus_area = random.choice(focus_areas)
    non_functionalities = random.sample(non_functionality_templates, k=3)

    # Return the generated values
    return {
        "service_name": service_name,
        "domain": domain,
        "platform": platform,
        "focus_area": focus_area,
        "functionalities": functionalities,
        "non_functionalities": non_functionalities
    }

generated_service = auto_generate_bfsi_service()
print(generated_service)


def generate_data(service_name, domain, platform, focus_area, functionalities, non_functionalities):
    prompt = f"""
    Generate structured data for the service '{service_name}'.
    - Domain: {domain}
    - Platform: {platform}
    - Focus Area: {focus_area}
    - Functionalities: {functionalities}
    - Non-Functionalities: {non_functionalities}
    Follow the specified format strictly, including additional components, errors/issues, and one good/bad user story with acceptance criteria.
    """
    
    response = openai.ChatCompletion.create(
        model="xyz",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=1500,
        temperature=0.7
    )
    return response.choices[0].message["content"]


generated_service = auto_generate_bfsi_service()
print(generated_service)
#extract data from generated_service to fill in data
data = generate_bfsi_data(service_name, domain, platform, focus_area, functionalities, non_functionalities)
```
